```python
import nltk
```

```python
from nltk.metrics.distance import jaccard_distance
from nltk.util import ngrams
```

```python
nltk.download('words')
from nltk.corpus import words
correct_words = words.words()
```

## True Words Non-Peaceful


```python
df0_list = df0["word"].tolist()

#wordcloud0 = WordCloud(width = 1000, height = 500).generate(df0_string)
```

```python
result=[]
for word in df0_list :
    temp = [(jaccard_distance(set(ngrams(word, 2)),
                              set(ngrams(w, 2))),w)
            for w in correct_words if w[0]==word[0]]
    result.append(sorted(temp, key = lambda val:val[0])[0][1])
```

```python
df0_string=(" ").join(result)
wordcloud0 = WordCloud(width = 1000, height = 500).generate(df0_string)
plt.figure(figsize=(15,8))
plt.imshow(wordcloud0)
plt.axis("off")
```

## True Words Peaceful

```python
df1_list = df1["word"].tolist()
result=[]
for word in df1_list :
    temp = [(jaccard_distance(set(ngrams(word, 2)),
                              set(ngrams(w, 2))),w)
            for w in correct_words if w[0]==word[0]]
    result.append(sorted(temp, key = lambda val:val[0])[0][1])
    
df1_string=(" ").join(result)
wordcloud1 = WordCloud(width = 1000, height = 500).generate(df1_string)
plt.figure(figsize=(15,8))
plt.imshow(wordcloud1)
plt.axis("off")
```
